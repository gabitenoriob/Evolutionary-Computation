```python
import random

def fitness(individuo):
    score = 0
    individuo_str = ''.join(map(str, individuo))
    for i in range(len(individuo_str) - 1):
        if individuo_str[i:i+2] == '01':
            score += 1
    return score

def crossover(mae, pai):
    ponto_cruzamento = random.randint(1, len(mae) - 1)
    print(f"Crossover point: {ponto_cruzamento}")
    filho = mae[:ponto_cruzamento] + pai[ponto_cruzamento:]
    print(f"Mae:   {mae}")
    print(f"Pai:   {pai}")
    print(f"Filho: {filho}")
    return filho

def mutacao(cromossomo, taxa=0.02):
    mutado = []
    for bit in cromossomo:
        if random.random() < taxa:
            novo_bit = 1 - bit
            print(f"Mutação: {bit} -> {novo_bit}")
            mutado.append(novo_bit)
        else:
            mutado.append(bit)
    return mutado

def ag():
    populacao = [[random.randint(0, 1) for _ in range(8)] for _ in range(10)]
    best, best_score = [], 0

    print("População inicial:")
    for ind in populacao:
        print(ind, "->", ''.join(map(str, ind)))

    for i in range(100):
        print(f"\n--- Geração {i+1} ---")

        for individuo in populacao:
            score = fitness(individuo)
            print(f"Indivíduo: {individuo}  ->  Score: {score}")

            if score > best_score:
                best, best_score = individuo[:], score
                print(f"Novo melhor indivíduo: {best} com score {best_score}")

            if best_score == 4:  # Máximo possível de "01" em 8 bits
                print("Solução ótima encontrada!")
                return best, best_score

        nova_populacao = []
        while len(nova_populacao) < len(populacao):
            pai = random.choice(populacao)
            mae = random.choice(populacao)
            print(f"\nSelecionados para crossover:\n  Mãe: {mae}\n  Pai: {pai}")
            filho = crossover(mae, pai)
            filho = mutacao(filho)
            nova_populacao.append(filho)

        populacao = nova_populacao
        print(f"Best individual so far: {best}, Score = {best_score}")

    return best, best_score


ag()
```
